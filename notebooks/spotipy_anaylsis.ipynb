```python
import pandas as pd

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from dotenv import load_dotenv

import sys
sys.path.append('..')

from src.database import TrackDatabase

db = TrackDatabase()
db.load()
```

```python
load_dotenv()

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials())

test_ids = db.dataset['track_id'].head(3).tolist()
results = sp.tracks(test_ids)

for track in results['tracks']:
    print(f"{track['name']} — {track['album']['release_date']}")
```
