```python
AUDIO_FEATURES = [
    'energy', 'danceability', 'valence', 'tempo',
    'acousticness', 'instrumentalness', 'speechiness', 
    'liveness', 'loudness'
]

# Compute mean of each feature per genre
genre_profiles = db.dataset.groupby('track_genre')[AUDIO_FEATURES].mean()

# Check it out
genre_profiles.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

# Normalizing the features
scaler = StandardScaler()
genre_profiles_normalized = pd.DataFrame(
    scaler.fit_transform(genre_profiles),
    index=genre_profiles.index,
    columns=genre_profiles.columns
)

genre_profiles_normalized.head(10)
```

```python
from scipy.spatial.distance import pdist, squareform

# Computing pairwise Euclidean distances between all genres
distances = pdist(genre_profiles_normalized, metric='euclidean')

# Converting to a square matrix (genre x genre)
distance_matrix = pd.DataFrame(
    squareform(distances),
    index=genre_profiles_normalized.index,
    columns=genre_profiles_normalized.index
)

# Checking distances from 'house' to other genres
distance_matrix['house'].sort_values().head(10)
```

```python
def explain_similarity(genre1: str, genre2: str, genre_profiles: pd.DataFrame, top_features: int = 2) -> str:
    """
    Explain why two genres are similar based on their closest features.
    """
    # Get raw feature values for both genres
    profile1 = genre_profiles.loc[genre1]
    profile2 = genre_profiles.loc[genre2]
    
    # Calculate absolute difference for each feature
    differences = abs(profile1 - profile2)
    
    # Get the features with smallest differences (most similar)
    closest_features = differences.sort_values().head(top_features).index.tolist()
    
    # Build explanation string
    explanations = []
    for feature in closest_features:
        val1 = profile1[feature]
        val2 = profile2[feature]
        
        # Format based on feature type
        if feature == 'tempo':
            explanations.append(f"{feature} ({val1:.0f} vs {val2:.0f} BPM)")
        elif feature == 'loudness':
            explanations.append(f"{feature} ({val1:.1f} vs {val2:.1f} dB)")
        else:
            explanations.append(f"{feature} ({val1:.2f} vs {val2:.2f})")
    
    return "Similar " + " and ".join(explanations)


# Test it
explain_similarity("house", "deep-house", genre_profiles)
```

```python
def get_similar_genres(
    genre: str, 
    distance_matrix: pd.DataFrame, 
    genre_profiles: pd.DataFrame,
    top_k: int = 10
) -> list:
    """
    Get the most similar genres with distance and explanation.
    """
    if genre not in distance_matrix.index:
        raise ValueError(f"Genre '{genre}' not found in dataset")
    
    # Get distances, sort, skip self (distance = 0)
    distances = distance_matrix[genre].sort_values()[1:top_k + 1]
    
    results = []
    for similar_genre, distance in distances.items():
        results.append({
            "genre": similar_genre,
            "distance": round(distance, 3),
            "explanation": explain_similarity(genre, similar_genre, genre_profiles)
        })
    
    return results


# Test it
get_similar_genres("house", distance_matrix, genre_profiles, top_k=5)
```
